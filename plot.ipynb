```python
import matplotlib.pyplot as plt
import numpy as np
import pickle as pk
MODELNAME = "Alpha20_8Head_128Hidden_Start_end"
META = f"loss_log/"+MODELNAME
with open(META,"rb") as f :
    meta = pk.load(f)
train_loss_log = meta['train_log']
val_loss_log = meta['val_log']
acc_log = meta['acc_log']
rc_log = meta['rc_log']
#train_path = "train_loss_100"
#val_path = "val_loss_100"
#with open(train_path, 'rb') as f:
#    train_loss_log = pk.load(f)
#with open(val_path, 'rb') as f:
#    val_loss_log = pk.load(f)
```

```python
x = np.arange(len(train_loss_log))
N = len(x)
#N = 50
plt.scatter(x[:N],train_loss_log[:N],label="Train")
plt.scatter(x[:N],val_loss_log[:N],label="Validatoin")
#plt.scatter(x,acc_log,label="Acc")
#plt.scatter(x,rc_log,label="Recall")
plt.xlabel("Epochs")
plt.ylabel("Cross Entropy Loss")
plt.title(MODELNAME)
plt.legend()
plt.show()
```

```python
#plt.scatter(x,train_loss_log,label="Train")
#plt.scatter(x,val_loss_log,label="Validatoin")
plt.scatter(x[:N],acc_log[:N],label="Acc")
plt.scatter(x[:N],rc_log[:N],label="Recall")
plt.xlabel("Epochs")
plt.ylabel("ACC and Recall")
plt.title(MODELNAME)
plt.legend()
plt.show()
```

```python
dum = np.array(val_loss_log)
print('Best Validation Loss:',np.argmin(dum))
dum = np.array(acc_log)
print('Best ACC:',np.argmin(dum))
dum = np.array(rc_log)
print('Best Recall:',np.argmin(dum))

```
